```python
import pandas as pd
import numpy as np
from nltk.tokenize import sent_tokenize, RegexpTokenizer
```

```python
test_eng = pd.read_parquet("../dataset/ML-ESG3_Testset_English.parquet")
test_fre = pd.read_parquet("../dataset/ML-ESG3_Testset_French_Translated.parquet")
test_kor = pd.read_parquet("../dataset/ML-ESG3_Testset_Korean_Translated.parquet")
```

```python
test_eng.head()
```

```python
test_eng["title_eng"] = test_eng["title"]
test_eng["content_eng"] = test_eng["content"]
test_eng["language"] = "English"
test_eng = test_eng[["title_eng", "content_eng", "language"]]
```

```python
test_fre.head()
```

```python
test_fre["language"] = "French"
test_fre = test_fre[["title_eng", "content_eng", "language"]]
```

```python
test_kor.head()
```

```python
test_kor["language"] = "Korean"
test_kor = test_kor[["title_eng", "content_eng", "language"]]
```

```python
df_test = pd.concat([test_eng, test_kor, test_fre]).reset_index(drop=True)
```

```python
regexp_tokenizer = RegexpTokenizer(r'\w+')

def word_count(article):
    return len(regexp_tokenizer.tokenize(article))
    

df_test["title_wc"] = df_test["title_eng"].apply(word_count)
df_test["content_wc"] = df_test["content_eng"].apply(word_count)

df_test[["title_wc", "content_wc"]].describe()
```

```python
# only korean dataset needs segmentation
print(df_test.groupby("language")["title_wc"].describe())
print(df_test.groupby("language")["content_wc"].describe())
```

```python
# Segment Chinese and Korean articles into every n sentences as they are too long

def group_sentences(sentences, sent_size):
    return [sentences[i : i + sent_size] for i in range(0, len(sentences), sent_size)]

def segment_articles(df, sent_size):
    df['sent_tokenize'] = df['content_eng'].apply(sent_tokenize)
    df['content_eng_short'] = df['sent_tokenize'].apply(lambda x: group_sentences(x, sent_size))
    seg_df = df.explode('content_eng_short')
    seg_df["content_eng_short"] = seg_df["content_eng_short"].apply(lambda x: " ".join(x))
    seg_df = seg_df.drop(columns="sent_tokenize")
    return seg_df
    
# segment Korean articles
sent_size = 5
test_kor_seg = segment_articles(test_kor, sent_size)
test_kor_seg["content_wc"] = test_kor_seg["content_eng_short"].apply(word_count)
```
